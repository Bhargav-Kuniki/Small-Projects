```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dxdt (y,x): 

    return y**0 + x

def dydt (y,x):

    return 10*y + 7*x

domain_start = -6
domain_end = 6
num = 37

x_domain = np.linspace(domain_start,domain_end,num)

y_domain = np.linspace(domain_start,domain_end,num)

#print(x_domain)
#print(y_domain)
```

```python
def slope_finder(input_domain,output_domain,func):

    slope_list = []
    a = 0

    while a < len(input_domain):

        slope_value = func(input_domain,output_domain[a])

        slope_list.append(slope_value)
        
        a = a + 1

    slope_array = np.array(slope_list)
    return slope_array

slope_array_x = slope_finder(x_domain,y_domain,dxdt)

slope_array_y = slope_finder(x_domain,y_domain,dydt)

#print(slope_array_x)
print('-------------')
#print(slope_array_y)
```

```python
x_comp = []
y_comp = []

for a in range (0,len(slope_array_x)):

    theta = np.arctan(slope_array_x[a]/slope_array_y[a])

    x_comp.append(np.cos(theta))
    y_comp.append(np.sin(theta))

x_comp = np.array(x_comp)
y_comp = np.array(y_comp)

```

```python
list = []

for x in range (0,num):

    list.append(y_domain)

list = np.array(list)

list_1 = []

for a in range(0,len(y_domain)):

    for b in range (0,len(y_domain)):

        list_1.append(list[a,b])

y_domain_graph = np.array(list_1)

#t_domain

list_2 = []

for a in range (0,num):

    intial = y_domain[a]

    b = 0
    while b < num: 

        list_2.append(intial)

        b = b + 1

x_domain_graph = np.array(list_2)


#print(y_domain_graph)
#print(t_domain_graph)
```

```python
fig, ax = plt.subplots()

ax.set_xlim([domain_start, domain_end])
ax.set_ylim([domain_start, domain_end])

#ax.axhline(0, color='black', linewidth=1)
#ax.axvline(0, color='black', linewidth=1)

ax.grid(True, linestyle = '--', alpha = 0.6)

ax.set_xlabel('x(t)')
ax.set_ylabel('y(t)')

ax.set_aspect('equal', adjustable='box')

def plot():

    plt.title('Phase Portrait')
    plt.quiver(x_domain_graph, y_domain_graph, y_comp, x_comp, color = 'black',headwidth=0, headlength=0, headaxislength=0 )
    #plt.plot(list_t,list_y)
    #plt.plot(list_t_neg, list_y_neg)
plot()
```
