```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# max_degree = highest degree of the functuion
# cof = coefficents of x^n in order of highest to lowest power
# x_0 = intial guess
# num_iter = number of times the program iterates


# Returns a graph of a polynomial given the maximum degree and coefficents of each term
# Plots a root
def graph_f_x(co,degree,x_domain,x_0):

    list_0 = []
    for a in range (len(x_domain)):
        x_value = x_domain[a]
        list_1 = []

        for b in range (degree + 1):
            i = b + 1
            value = co[-i]*(x_value)**b
            list_1.append(value)

        list_0.append(np.sum(list_1))

    def plot(x_val,y_val):

        plt.plot(x_val,y_val)
        plt.plot(x_0,0,'o')
        plt.axhline(0, color='gray', linestyle='--', linewidth=1.5)
        plt.xlabel('x')
        plt.ylabel('y')
        plt.show
    plot(x_domain,list_0)


def newtons_method_1(max_degree,cof,x_0,num_iter):

    def power_poly_prime(max_degree,cof,x_val):

        power_list = np.linspace(max_degree,0,max_degree + 1)
        
        prime_cof = power_list * cof
        prime_power = power_list[1:len(power_list)]

        lst_prime = []
        lst_power = []

        for a in range (len(prime_power)):
            co = prime_cof[a]
            power = prime_power[a]

            value = co * (x_val ** power)
            lst_prime.append(value)

        for b in range (len(power_list)):

            co = cof[b]
            power = power_list[b]

            value = co * (x_val ** power)
            lst_power.append(value)

        return sum(lst_power), sum(lst_prime)

    for a in range (num_iter):

        f_x, f_x_prime = power_poly_prime(max_degree,cof,x_0)
        x_1 = x_0 - f_x/f_x_prime
        
        x_0 = x_1

    return x_1

x_domain = np.linspace(-4.5,1,150)
deg = 4
cof = [1,5,3,-2,-1]



a = newtons_method_1(deg,cof,1,200)
print(a)

graph_f_x(cof,deg,x_domain,a)

```
