```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
def f_x(x):
    return 1/x

lower_bound = 1
upper_bound = np.exp(5)
number_of_terms = 10**7

# If you wanna check the error enter a known value for you integral

real_value = 5
number = 100
```

```python
# definite integral calculator where the b - a is small 

def int_small_calc(a,b,acc):

    domain = np.linspace(a,b,acc)
    delta_x = (b - a)/acc

    f_x_array_1 = f_x(domain)

    sum = f_x_array_1 * delta_x

    int_sum_mid = np.sum(sum)

    return int_sum_mid

value = int_small_calc(lower_bound,upper_bound,number_of_terms)

print(value)
```

```python
#definite integral calculator where the b - a is large

def int_large_calc(a,b,acc):


    domain_1 = np.linspace(a,b/2,acc)
    domain_2 = np.linspace(b/2,b,acc)

    delta_x = (b - a)/acc

    f_x_array_1 = f_x(domain_1) * delta_x
    f_x_array_2 = f_x(domain_2) * delta_x


    sum = f_x_array_1 + f_x_array_2

    int_sum_mid = np.sum(sum)

    return int_sum_mid/2

value = int_large_calc(lower_bound,upper_bound,number_of_terms)

print(value)
```

```python
x = 1
list_1 = []
list_2 = []

while x <= number : 

    acc_num = x


    approx_value = int_small_calc(lower_bound,upper_bound,acc_num)

#Err percent
    #diff = np.abs(approx_value - real_value)*100/(real_value)

#Err differance
    diff = np.abs(approx_value - real_value)*100/(real_value)


    list_1.append(diff)
    list_2.append(acc_num)

    x = x + 1

diff_arr = np.array(list_1)
number_term_arr = np.array(list_2)
```
